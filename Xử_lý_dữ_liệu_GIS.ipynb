```python
import geopandas as gpd
```

```python
df = gpd.read_file('/content/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_Ward_Level.shp')
```

```python
df
```

```python
#phường có diện tích lớn nhất
max_area = df['Shape_Area'].max()
maxarena = df[['Com_Name', 'Dist_Name', 'Shape_Area']][df['Shape_Area'] == max_area]
print("Phường có diện tích lớn nhất là phường "+maxarena['Com_Name'].to_string(index=False)
+" quận " +maxarena['Dist_Name'].to_string(index=False))
```

```python
maxarena
```

```python
#phường có dân số 2019 (Pop_2019) cao nhất
max_pop_2019 = df['Pop_2019'].max()
maxpop2019 = df[['Com_Name', 'Dist_Name', 'Pop_2019']][df['Pop_2019'] == max_pop_2019]
print("Phường có có dân số 2019 (Pop_2019) cao nhất là phường "+maxpop2019['Com_Name'].to_string(index=False)
+" quận " +maxpop2019['Dist_Name'].to_string(index=False))
```

```python
maxpop2019
```

```python
#phường có diện tích nhỏ nhất
min_area = df['Shape_Area'].min()
minarea = df[['Com_Name', 'Dist_Name', 'Shape_Area']][df['Shape_Area'] == min_area]
print("Phường có diện tích nhỏ nhất là phường "+minarea['Com_Name'].to_string(index=False)
+" quận " +minarea['Dist_Name'].to_string(index=False))
```

```python
minarea
```

```python
#phường có dân số 2019 (Pop_2019) nhỏ nhất
min_pop_2019 = df['Pop_2019'].min()
minpop2019 = df[['Com_Name', 'Dist_Name', 'Pop_2019']][df['Pop_2019'] == min_pop_2019]
print("Phường có có dân số 2019 (Pop_2019) nhỏ nhất là phường "+minpop2019['Com_Name'].to_string(index=False)
+" quận " +minpop2019['Dist_Name'].to_string(index=False))
```

```python
minpop2019
```

```python
#phường có tốc độ tăng trưởng dân số nhanh nhất (dựa trên Pop_2009 và Pop_2019)
df['Toc_do'] = df['Pop_2019'] / df['Pop_2009']
max_tocdo = df['Toc_do'].max()
max_tocdo = df[['Com_Name', 'Dist_Name','Toc_do']][df['Toc_do'] == max_tocdo]
print("Phường có tốc độ tăng trưởng dân số nhanh nhất (dựa trên Pop_2009 và Pop_2019) là phường "+max_tocdo['Com_Name'].to_string(index=False)
+" quận " +max_tocdo['Dist_Name'].to_string(index=False))
```

```python
max_tocdo
```

```python
#phường có tốc độ tăng trưởng dân số thấp nhất
df['Toc_do'] = df['Pop_2019'] / df['Pop_2009']
min_tocdo = df['Toc_do'].min()
min_tocdo = df[['Com_Name', 'Dist_Name','Toc_do']][df['Toc_do'] == min_tocdo]
print("Phường có tốc độ tăng trưởng dân số thấp nhất là phường "+min_tocdo['Com_Name'].to_string(index=False)
+" quận " +min_tocdo['Dist_Name'].to_string(index=False))
```

```python
min_tocdo
```

```python
#phường có biến động dân số nhanh nhất
df['Bien_dong'] = df['Pop_2019'] - df['Pop_2009']
max_biendong = df['Bien_dong'].max()
max_biendong = df[['Com_Name', 'Dist_Name','Bien_dong']][df['Bien_dong'] == max_biendong]
print("Phường có biến động dân số nhanh nhất là phường "+max_biendong['Com_Name'].to_string(index=False)
+" quận " +max_biendong['Dist_Name'].to_string(index=False))
```

```python
max_biendong
```

```python
#phường có biến động dân số chậm nhất
df['Bien_dong'] = df['Pop_2019'] - df['Pop_2009']
min_biendong = df['Bien_dong'].min()
min_biendong = df[['Com_Name', 'Dist_Name','Bien_dong']][df['Bien_dong'] == min_biendong]
print("Phường có biến động dân số chậm nhất là phường "+min_biendong['Com_Name'].to_string(index=False)
+" quận " +min_biendong['Dist_Name'].to_string(index=False))
```
